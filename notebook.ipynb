## Beta

```python
def calculate_beta(data):
    stock_returns = data['Close']
    market_returns = data['Open']
    beta = np.cov(market_returns, stock_returns, ddof=0)[0][1] / np.var(market_returns, ddof=0)
    return beta

# Assuming df_BNP, df_MC, and df_TTE are your DataFrames
beta_BNP = calculate_beta(df_BNP)
beta_MC = calculate_beta(df_MC)
beta_TTE = calculate_beta(df_TTE)

print("Beta")
print("BNP:", beta_BNP)
print("MC:", beta_MC)
print("TTE:", beta_TTE)  
```

## Cac40 Index Hedge

```python
def calculate_cac40_hedge(data):
    cac40_index = data[['Open']].mean(axis=1)
    beta = np.cov(cac40_index, data['Open'], ddof=0)[0][1] / np.var(cac40_index, ddof=0)
    return beta
```

```python
beta_BNP = calculate_cac40_hedge(df_BNP)
beta_MC = calculate_cac40_hedge(df_MC)
beta_TTE = calculate_cac40_hedge(df_TTE)

print("Beta against CAC40 Index")
print("BNP:", beta_BNP)
print("MC:", beta_MC)
print("TTE:", beta_TTE)
```

```python
def find_outliers_iqr(data, column_name):
    # Calculate quartiles and IQR for the specified column
    Q1 = data[column_name].quantile(0.25)
    Q3 = data[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Calculate the lower and upper bounds to identify outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Find outliers in the specified column
    outliers = data[(data[column_name] < lower_bound) | (data[column_name] > upper_bound)]
    
    return outliers
```

```python
def create_boxplot_with_outliers(data, column_name):
    # Finding outliers in the specified column
    outliers = find_outliers_iqr(data, column_name)

    # Creating a boxplot to visualize the specified column and its outliers
    plt.figure(figsize=(8, 6))
    plt.boxplot(data[column_name])
    plt.title(f'Boxplot of "{column_name}" Column with Outliers')
    plt.ylabel('Values')
    plt.xlabel(column_name)
    plt.grid(True)

    # Highlighting outliers on the boxplot
    if not outliers.empty:
        plt.scatter(outliers.index + 1, outliers[column_name], color='red', label='Outliers')
        plt.legend()

    plt.show()
```

```python
def visualize_histogram(data, column_name):
    plt.figure(figsize=(8, 6))
    sns.histplot(data=data[column_name], kde=True, color='navy')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of "{column_name}" Column')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

def visualize_dataframe_histogram(data):
    plt.figure(figsize=(10, 6))
    plt.title('Histogram of DataFrame Columns')

    for col in data.columns:
        if data[col].dtype != 'datetime64[ns]':  # Exclude datetime columns
            plt.hist(data[col], bins=20, alpha=0.5, label=col)

    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
    
def visualize_boxplot(data, column_name):
    plt.figure(figsize=(6, 8))
    plt.boxplot(data[column_name])
    plt.title(f'Boxplot of "{column_name}" Column')
    plt.ylabel('Values')
    plt.xlabel(column_name)
    plt.show()
    
def visualize_time_series(data, x_column, y_columns):
    plt.figure(figsize=(10, 6))
    data[x_column] = pd.to_datetime(data[x_column])  # Convert 'date' column to datetime
    deciles = np.percentile(data[x_column].dt.month, np.arange(10, 101, 10))  # Calculate deciles by month
    
    for column in y_columns:
        plt.plot(data[x_column], data[column], label=column)
    
    plt.title('Time Series Plot')
    plt.xlabel('Month')
    plt.ylabel('Value')
    plt.legend()
    
    # Set x-axis ticks to month names corresponding to deciles
    month_names = [calendar.month_name[i] for i in range(1, 13)]
    decile_month_names = [month_names[int(d) - 1] for d in deciles]
    
    plt.xticks(ticks=pd.to_datetime(deciles), labels=decile_month_names)
    
    plt.tight_layout()
    plt.show()

    
def visualize_correlation_matrix(data):
    plt.figure(figsize=(10, 8))
    sns.set(style='white')  # Set style for the plot
    
    # Calculate correlation matrix
    correlation_matrix = data.corr()
    
    cmap = np.random.choice(plt.colormaps())
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap=cmap, fmt='.2f')
    plt.title('Correlation Matrix Heatmap')
    plt.show()
```

## Variance

```python
def calculate_column_variance(data, column_name):
    column_variance = data[column_name].var()
    return column_variance

def calculate_column_covariance(data, column):
    numeric_columns = data.select_dtypes(include=[np.number]).columns.drop(column)
    covariances = {col: data[[column, col]].cov().iloc[0, 1] for col in numeric_columns}
    print(f"Covariance of '{column}' with other columns:")
    for col, cov in covariances.items():
        print(f"{col}: {cov:.4f}")
```

## Portfolio VaR

```python
def portfolio_value_at_risk(data, num_assets, confidence_level=0.95):
    weights = np.array([1 / num_assets] * num_assets)  # Equal weights for all assets
    portfolio_returns = np.dot(data.iloc[:, 1:].pct_change().dropna(), weights)
    portfolio_value_at_risk = np.percentile(portfolio_returns, 100 * (1 - confidence_level))
    return portfolio_value_at_risk
```
